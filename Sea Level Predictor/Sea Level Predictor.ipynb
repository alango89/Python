```python
#Use the linregress function from scipy.stats to get the slope and y-intercept of the line of best fit. 
regresion = linregress(df['Year'], df['CSIRO Adjusted Sea Level'])
```

```python
#Plot the line of best fit over the top of the scatter plot.
plt.figure(figsize=(16,9))
plt.scatter(df['Year'], df['CSIRO Adjusted Sea Level'])
plt.plot(df['Year'], regresion.intercept + regresion.slope*df['Year'], label='fitted line', color='gold')
plt.legend()
plt.title('Rise in Sea Level')
plt.xlabel('Year')
plt.ylabel('Sea Level (inches)')
plt.show
```

```python
#Make the line go through the year 2050 to predict the sea level rise in 2050.
year_extended = list(range(1880, 2050, 1))
line = [regresion.intercept + regresion.slope * j for j in year_extended]
plt.figure(figsize=(16,9))
plt.scatter(df['Year'], df['CSIRO Adjusted Sea Level'])
plt.plot(year_extended, line, linewidth=2, color="r")
plt.title('Rise in Sea Level')
plt.xlabel('Year')
plt.ylabel('Sea Level (inches)')
plt.show
```

```python
#Plot a new line of best fit just using the data from year 2000 through the most recent year 
df_new = df.loc[(df['Year'] >= 2000) & (df['Year'] <= 2023)]
regresion2 = linregress(df_new['Year'], df_new['CSIRO Adjusted Sea Level'])

year_extended = list(range(2000, 2050, 1))
line = [regresion2.intercept + regresion2.slope * j for j in year_extended]
plt.figure(figsize=(16,9))
plt.scatter(df_new['Year'], df_new['CSIRO Adjusted Sea Level'])
plt.plot(year_extended, line, linewidth=2, color="gold")
plt.title('Rise in Sea Level')
plt.xlabel('Year')
plt.ylabel('Sea Level (inches)')
plt.show
```
